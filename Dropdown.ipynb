```python
import dash
from dash import dcc,html,Input,Output
import plotly.express as px
import pandas as pd
data =pd.read_csv('/content/output.csv')
data.head()
```

```python
# prompt: generate a code for dropdown from above data

import dash
from dash import dcc,html,Input,Output
import plotly.express as px
import pandas as pd

# Assuming 'data' DataFrame is already loaded from the previous code

app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Dropdown Example"),
    dcc.Dropdown(
        id='dropdown',
        options=[{'label': i, 'value': i} for i in data.columns], # Create options from DataFrame columns
        value=data.columns[0]  # Set initial value
    ),
    dcc.Graph(id='graph')
])

@app.callback(
    Output('graph', 'figure'),
    Input('dropdown', 'value')
)
def update_graph(column_name):
    fig = px.histogram(data, x=column_name)
    return fig


if __name__ == '__main__':
    app.run_server(debug=True)
```
