```python
# prompt: generate a code to fill the missing values in Make ,and Colour with mode values

import pandas as pd
import numpy as np

data = pd.read_csv('/content/Car Sales Missing Data.csv')

# Fill missing 'Make' values with the mode
make_mode = data['Make'].mode()[0]  # Get the most frequent 'Make'
data['Make'].fillna(make_mode, inplace=True)

# Fill missing 'Colour' values with the mode
colour_mode = data['Colour'].mode()[0]  # Get the most frequent 'Colour'
data['Colour'].fillna(colour_mode, inplace=True)

print(data.isnull().sum()) #check if the null values are filled
data.tail()
```

```python
# prompt: generate a code to fill missing values in Odometer (KM) with mean , median, mode

import pandas as pd
import numpy as np

data = pd.read_csv('/content/Car Sales Missing Data.csv')

# Fill missing 'Odometer (KM)' values with the mean
odometer_mean = data['Odometer (KM)'].mean()
data['Odometer (KM)'].fillna(odometer_mean, inplace=True)

# Create a copy for median filling
data_median = data.copy()
odometer_median = data_median['Odometer (KM)'].median()
data_median['Odometer (KM)'].fillna(odometer_median, inplace=True)

# Create a copy for mode filling
data_mode = data.copy()
odometer_mode = data_mode['Odometer (KM)'].mode()[0] # Mode might return multiple values, take the first one
data_mode['Odometer (KM)'].fillna(odometer_mode, inplace=True)

# Display the filled dataframes
print("Original Data with Mean Imputation:")
print(data.tail())
print("\nOriginal Data with Median Imputation:")
print(data_median.tail())
print("\nOriginal Data with Mode Imputation:")
print(data_mode.tail())
```

```python
# prompt: generate a code to fill the missing values in Doors with mean , median , mode values

import pandas as pd
import numpy as np

data = pd.read_csv('/content/Car Sales Missing Data.csv')

# Fill missing 'Doors' values with the most frequent value (mode)
doors_mode = data['Doors'].mode()[0]
data['Doors'].fillna(doors_mode, inplace=True)

# Calculate mean, median, and mode for demonstration
doors_mean = data['Doors'].mean()
doors_median = data['Doors'].median()
doors_mode = data['Doors'].mode()[0]

print(f"Mean of 'Doors': {doors_mean}")
print(f"Median of 'Doors': {doors_median}")
print(f"Mode of 'Doors': {doors_mode}")

print("\nData with missing 'Doors' filled using mode:")
print(data.tail())
```
