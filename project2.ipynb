```python
# prompt: generate a code to Create a new column for the age of the car using the registration field

import pandas as pd

# Assuming 'data_cleaned' DataFrame is available from the previous code.
# If not, load and clean the data as shown in the previous code.

# Calculate the current year
current_year = pd.Timestamp.now().year

# Function to calculate car age
def calculate_car_age(registration_date):
    if pd.isnull(registration_date):
        return None  # Handle missing registration dates
    else:
        return current_year - registration_date.year


# Apply the function to create the 'car_age' column
data_cleaned['car_age'] = data_cleaned['registration'].apply(calculate_car_age)

# Print the updated DataFrame to verify the new column
print(data_cleaned.head())
```
